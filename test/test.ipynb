# 打标准确率计算

```python
import json
import pandas as pd

count = 50

result = pd.read_excel('PR外投跟进数据_AI.xlsx').iloc[:count]
result['外投问题类型(一级)'] = result['外投问题类型(一级)'].map(lambda x: json.loads(x.replace('\'', '"')))
result['外投问题类型(一级)_AI'] = result['外投问题类型(一级)_AI'].map(lambda x: json.loads(x.replace('\'', '"')))
result['外投问题类型(二级)'] = result['外投问题类型(二级)'].map(lambda x: json.loads(x.replace('\'', '"')))
result['外投问题类型(二级)_AI'] = result['外投问题类型(二级)_AI'].map(lambda x: json.loads(x.replace('\'', '"')))
```

```python
def get_correct_point(l1, l2):
    s1, s2 = set(l1), set(l2)
    if s1 | s2 == s1 and s1 | s2 == s2:
        return 1
    if s1 & s2 == set():
        return 0
    return 0.5

def get_precision(l1, l2):
    l2_correct = [i for i in l2 if i in l1]
    return len(l2_correct)
        
```

```python
# 一级问题准确率
cor = result.iloc[:].apply(lambda ser: get_correct_point(ser['外投问题类型(一级)'], ser['外投问题类型(一级)_AI']), axis=1)
cor.sum() / count
```

```python
# 二级问题准确率
cor = result.apply(lambda ser: get_correct_point(ser['外投问题类型(二级)'], ser['外投问题类型(二级)_AI']), axis=1)
cor.sum() / count
```

```python
# 一级问题 micro precision

level_1_correct_cnt_sum = result.apply(lambda ser: get_precision(ser['外投问题类型(一级)'], ser['外投问题类型(一级)_AI']), axis=1).sum()
level_1_cnt_sum = result['外投问题类型(一级)_AI'].map(len).sum()

micro_precision = level_1_correct_cnt_sum / level_1_cnt_sum
micro_precision
```
